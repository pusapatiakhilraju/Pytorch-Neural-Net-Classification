# Finding the min, max, mean, sum (Tensor Aggregation)

```python
x = torch.arange(1,100,10)
# Find the minimum
x.min(), torch.min(x)
# Find the maximum
x.max(), torch.max(x)
# Find the average
x.type(torch.float32).mean(), torch.mean(x.type(torch.float32))
# find the position of the min and max values 
x.argmin(), x.argmax()
```

# Reshaping, Stacking, Squeezing, Unsqeezing Tensors

```python
x = torch.arange(1.0, 10.0 , 1)
# Reshape
x_reshaped = torch.reshape(x,(-1,))
# View
x_view = x.view((-1,1))
# Squeze
x_squeezed = torch.squeeze(x, dim = 0)
# unsqueze
x_unsqueezed = torch.unsqueeze(x, dim= 0)
# permute
image = torch.rand(224,224,3)
x_permuted = torch.permute(image,(2,0,1))

```
